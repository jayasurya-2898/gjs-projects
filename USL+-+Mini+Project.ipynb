```python
for i in df_num_float.columns:
    sb.boxplot(df_num_float[i])
    plt.show()
```

```python
for i in df_num_float.columns:
    q1 = df_num_float[i].quantile(.25)
    q3 = df_num_float[i].quantile(.75)
    iqr = q3-q1
    ub = q3 + (1.5* iqr)
    lb = q1 - (1.5* iqr)
    
    for j in df_num_float.index:
        if df_num_float.loc[j,i]> ub:
            df_num_float.loc[j,i]= ub
        if df_num_float.loc[j,i]< lb:
            df_num_float.loc[j,i]= lb
```
